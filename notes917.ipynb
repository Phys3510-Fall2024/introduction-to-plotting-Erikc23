```python
#importing numpy and matplot
import numpy as np
import matplotlib.pyplot as plt
```

```python
#def U
def pot_E(x,k = 1.0):
    return k*x**2
```

```python
x = np.linspace(-5, 5, 100)
y =  pot_E(x)
plt.plot(x,y,color="black", linewidth=0.5)
plt.xlabel("displacement")
plt.ylabel("energy")
plt.xlim([-4,4])
plt.ylim([-1,10])
plt.show()
```

```python
def pote2d(x,y,kx=1.0,ky=1.00):
    return 0.5*(kx*x**2 +ky *y**2)
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
X,Y=np.meshgrid(x,y)
Z=pote2d(X,Y)
#print(X)
#print()
#print(Y)
#print()
#print(Z)
```

```python
#cont plot
plt.contour(X,Y,Z)
```

```python
#cont plot filled
plt.contourf(X,Y,Z)
```

```python
plt.contourf(X,Y,Z)
#title 
plt.title("Title")
#x axis labe
plt.xlabel("x displacement")
#y axis label
plt.ylabel("y-displacement")
#adds legend
plt.colorbar()
#limits of graph
plt.xlim([-5,5])
plt.ylim([-5,5])
#show graph
plt.show()
```
